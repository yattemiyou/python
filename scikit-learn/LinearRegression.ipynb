# 線形回帰（単回帰）

```python
import japanize_matplotlib
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score

# 学習データ（y = 2x + 1）
X = [[0], [1], [2], [3]]
y = [1, 3, 5, 7]

# 線形回帰モデル
model = LinearRegression()

# 学習
model.fit(X, y)

# 学習結果
a = model.coef_[0]
b = model.intercept_
print(f'学習結果：y = {a} * x + {b}')

# 検証
y_pred = model.predict(X)
print(f'検証結果：{accuracy_score(y, y_pred)}')

# 予測
data = [[0.75], [1.3], [2.8]]
print(f'予測結果：{model.predict(data)}')

'''
可視化
'''
# 回帰直線をプロット
x = np.arange(0, 5, 1)
plt.plot(x, a*x+b, label=f'y = {a} * x + {b}')

# 学習データをプロット
plt.plot(X, y, 'bo', label='学習データ')

# 予測データをプロット
plt.plot(data, model.predict(data), 'ro', label='予測データ')

plt.title('線形回帰（単回帰）')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
